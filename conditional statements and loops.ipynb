### if statement
* The if statement is the most simple decision-making statement. It is used to decide whether a certain statement or block of statements will be executed or not.
* Syntax:
* if (condition):
1) Executes this block if
2) condition is true
* else:
1) Executes this block if
2) condition is false

```python
sathwika_marks=98
if sathwika_marks>=80: # no white spaces should involve
    print('top scorer') #  the output returns true
```

```python
# by using if statement
num= int(input())
if num%12 == 0: # if it true goes to the next statement
    if num%6 == 0: # percentile symbol is denoted as reminder
        print('div by 24')
```

### If-Else Statement
* if statement alone tells us that if a condition is true it will execute a block of statements and if the condition is false it wont.
* But if we want to do something else if the condition is false, we can use the else statement with if statement to execute a block of code when the if condition is false.
* Syntax: if (condition):
1) Executes this block if
2) condition is true
* else:
1) Executes this block if
2) condition is false

```python
x = 16
if x == 17:      
   print("Yes")
else:
   print("No")
```

### Nested if Statement
* You can have if statements inside if statements, this is called a nested if.
* This means that inner if condition will be checked only if outer if condition is true and by this, we can see multiple conditions to be satisfied.
* Syntax: 
* if (condition1):
1) Executes if condition 1 is true
* if (condition2):
1) Executes if condition 1 and condition 2 is true
2) if Block is end here

```python
# using nested if statement
num= int(input())
# merging two conditions based on (and gate & or gate)
if num%12 == 0 or num%6 == 0:
        print('div by 24')
```

### if-elif Statement
* if-elif statement is shortcut of if..else chain.
* While using if-elif statement at the end else block is added which is performed if none of the above if-elif statement is true.
* Syntax:-
* if (condition):
1) statement
* elif (condition):
1) statement
* else:
1) statement

```python
# above 80 --> second round
# 70 - 80 --> first round
# 60 - 70 --> screening round
# below 60 --> study hard

marks = int(input('enter your marks: ')) #by using integers values
if marks >=80:        # followed a condition here marks
    print('second round')
elif marks >= 70:    # if marks in between 70 to 79 
    print('first round') # first round will be printed
elif marks >=60:
    print('screening round')
else:             # written outside the if
    print('study hard')
```

## 2) iterative statements in python
* Iteration statements or loop statements allow us to execute a block of statements as long as the condition is true.
* Loops statements are used when we need to run same code again and again, each time with a different value.
### types of iteration statements in python
* In Python Iteration (Loops) statements are of three types :-
1) while loop
2) for loop
3) nested for loops (not primitive)

### while loop in python
* While Loop In Python is used to execute a block of statement as long as a given condition is true.
* when the condition is false, the control will come out of the loop.
* The condition is checked every time at the beginning of the loop.
* whileloop is the repetation of the if
* syntax:
1) while (condition):
   * statements
* Python uses indentation as its method of grouping statements.
* While loop falls under the category of indefinite iteration. Indefinite iteration means that the number of times the loop is executed isnt specified explicitly in advance.
* Statements represent all the statements indented by the same number of character spaces after a programming construct are considered to be part of a single block of code.
* Python uses indentation as its method of grouping statements.
* When a while loop is executed, expr is first evaluated in a Boolean context and if it is true, the loop body is executed.
* Then the expr is checked again, if it is still true then the body is executed again and this continues until the expression becomes false.

### enter while loop:-


```python
# example of while loop
x = 2
while (x<5): # printing the values below 5 starts from x value
    print(x)
    x += 1
# Note: remember to increment x, or else the loop will continue forever.
```

```python
num1=int(input('y:'))
num2=num1+10
while num1 < num2:
    num1+=1
    print(num1) # printed the of y 6 then added +10.
```

### for loop in python
* for loop starts end of the iterator
* For loop in Python is used to iterate over items of any sequence, such as a list, a tuple, a dictionary, a set, or a string.
* This is less like the for keyword in other programming languages, and works more like an iterator method as found in other object-orientated programming languages.
* With the for loop we can execute a set of statements, once for each item in a list, tuple, set etc.
* note : The for loop doesn't require an indexing variable to set beforehand.
* syntax : 
1) for val in sequence:
   * statements

#### for loop 


```python
# To loop through a set of code a specified number of times
for i in range (1,15,3): # in range of 1 to 15, every 3 values are removed.
    print (i)
```

### Loop Control Statements
* Loop control statements change execution from their normal sequence.
* When execution leaves a scope, all automatic objects that were created in that scope are destroyed.
* Python supports the following control statements.
1) Pass Statement
2) Continue Statement
3) Break Statement

#### 1) pass statement
* it simply does nothing
* pass is a placeholder(it be there but don't know what to code)
* it is used when a statement is required syntactically but we don't want any command or code to execute.
* it is like a null operator (nothing happens even if executed)
* pass statement is used for writing empty loos.
* pass statement is also used for empty control statement, functions and classes.
* note : Note : Pass statement creates a problem inside loop while as it keeps looping if the given condition is true.
* syntax : 
1. function/ condition / loop:
   * pass

```python
# for loop through pass statement
v =['w', 'x', 'y', 'z']
for i in v:
    if(i =='w'):
        pass
    else:
        print(i)
```

#### 2) continue statement
* The Python continue statement is used to skip the next part of the code in the iteration and restart with the next iteration of the loop. (it encounters restart the loop)
* In an iteration, if we want to skip a certain iteration based on the condition specified and then carry on with the further code, we make use of continue statement.
* return : the control to the beginning of the current loop. When encountered, the loop starts next iteration without executing the remaining statements in the current iteration.
* The continue statement can be used in both while and for loops.
* The continue statement is just the opposite to that of break. It skips the remaining statements in the current loop and starts the next iteration.

```python
# for loop throuh continue statement 
for i in range(1,30,3):
    # if i equals to 2, # removes every 2 values in the range
    # continue to next iteration
    # without printing
    if i == 2:
        continue
    else:
            # otherwise print the value of i
            print(i, sep = '\n')
```
